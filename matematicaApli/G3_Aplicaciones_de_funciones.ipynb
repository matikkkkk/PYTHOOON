# Problema 1

Sean las funciones:

\begin{split}
r(x) & = -1,5x^{2} + 11,5x - 15\\
g(x) & = 1,58x^{4} - 19,17x^{3} + 80,92x^{2} - 139,33x + 85
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $0 \leq x \leq 6$, $\hspace{0.3cm}$  $-10 \leq y \leq 10$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones considerando el dominio indicado. Sugerencia: utilice *plt.xlim()* y *plt.ylim()* para definir los límites de los ejes en el gráfico.
2. ¿Cuáles son los puntos de intersección entre las funciones? Utilice *fsolve* para encontrar las soluciones.

1. Grafique ambas funciones considerando el dominio indicado. Sugerencia: utilice *plt.xlim()* y *plt.ylim()* para definir los límites de los ejes en el gráfico.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
def r(x):
    return -1.5*x**2+11.5*x-15
def g(x):
    return 1.58*x**4-19.17*x**3+80.92*x**2-139.33*x+85
x = np.arange(0,6.1,0.1)
plt.plot(x, r(x), label = 'r(x)')
plt.plot(x, g(x), label = 'g(x)')
plt.axis([0,6,-10,10])
plt.title('Nombre del gráfico')
plt.xlabel("eje x")
plt.ylabel('Nombre eje y')
plt.grid(True)
plt.legend()
plt.show()
```

2. ¿Cuáles son los puntos de intersección entre las funciones? Utilice *fsolve* para encontrar las soluciones.

```python
def L(x):
    return r(x)-g(x)
xo = np.linspace(1.5,5.5,100)
solucion = np.around(fsolve(L, xo),4)
print(solucion)
sol_unicas=np.unique(solucion)
print(sol_unicas)
```

X=2.2282

X=2.8814

X=5.0845

# Problema 3

Sean las funciones:

\begin{split}
w(x) & = 0,6x^{2} + 1,5\\
t(x) & = 0,8x^{3} - 10x + 10
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $-3 \leq x \leq 4$, $\hspace{0.2cm}$ $-4 \leq y \leq 24$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones en un mismo gráfico, indicando unidad de medida. Utilice *Matplotlib*.
2. Determine el intervalo en que $w(x) \geq t(x)$.
3. Encuentre los puntos en que $t(x) = 6$.


1. Grafique ambas funciones en un mismo gráfico, indicando unidad de medida. Utilice *Matplotlib*.


```python
def w(x):
    return 0.6*x**2+1.5
def t(x):
    return 0.8*x**3-10*x+10
x = np.arange(-3,5,0.1)
plt.plot(x, w(x), label = 'w(x)')
plt.plot(x, t(x), label = 't(x)')
plt.axis([-3,4,-4,25])
plt.title('Nombre del gráfico')
plt.xlabel("eje x")
plt.ylabel('Nombre eje y')
plt.grid(True)
plt.legend()
plt.show()
```

2. Determine el intervalo en que $w(x) \geq t(x)$.


```python
def K(x):
    return w(x)-t(x)
xo = np.linspace(0,5,2)
solucion = np.around(fsolve(K, xo),2)
print(solucion)
print(F"se intersecan en los puntos {solucion}")
```

Por lo tanto el intervalo es entre 0.86 y  3.47

3. Encuentre los puntos en que $t(x) = 6$.


```python
def A(x):
    return t(x)-6
xo = np.linspace(0,10,1)
solucion = fsolve(A, xo)
print(solucion)
```
