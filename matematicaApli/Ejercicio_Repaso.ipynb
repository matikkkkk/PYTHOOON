3. Calcule el tiempo que tarda la impresora en procesar 17 hojas.

```python
def f(x):
    return 2.5*x+6
print(f" el tiempo que demora la impresora en procesar 17 hojas es de {f(17)} segundos ")
```

4. Si la impresora tardó 2 minutos y 6 segundos, ¿Cuántas hojas imprimió?

```python
def A(x):
    return f(x)-126
xo = np.linspace(0,20,100)
soluciones = np.around(fsolve(A, xo))
soluciones_unicas = np.unique(soluciones)
print(soluciones_unicas)
```

5. El jefe nos indica que hay otra impresora que podría ser conveniente. Ésta tarda 1,5 segundos por hoja, pero toma un tiempo de 25 segundos en
prenderse. ¿Para cuántas hojas es más conveniente esta segunda impresora?

```python
def O(x):
    return 1.5*x+25
```

```python
x = np.arange(0,50,0.01)
plt.plot(x, f(x), label = 'f(x)')
plt.plot(x, O(x), label = 'A(x)')
plt.axis([1,30,7.5,100])
plt.title('Nombre del gráfico')
plt.xlabel('Nombre eje x')
plt.ylabel('Nombre eje y')
plt.grid(True)
plt.legend()
plt.show()
```

```python

def Z(x):
    return f(x)-O(x)
xo = np.linspace(15,30,100)
soluciones = np.around(fsolve(Z, xo),2)
soluciones_unicas = np.unique(soluciones)
print(soluciones_unicas)
```
