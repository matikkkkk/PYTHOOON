Un equipo de desarrollo está evaluando la eficiencia de un nuevo algoritmo implementado en *Python*. Han medido el tiempo de ejecución del algoritmo en función de la cantidad de elementos de entrada, obteniendo los resultados que se muestran en la siguiente tabla:

<p><center>

| Cantidad de elementos (unidades) | Tiempo de ejecución (ms) |
|----------------------------------|--------------------------|
|$$ 100 $$                         | $$2 $$                   |
|$$ 200 $$                         | $$4 $$                   |

|$$ 1.000 $$                        | $$20 $$                  |
|$$ 2.000 $$                        | $$40 $$                  |

</center></p>

1. Defina variable dependiente e independiente, indicando unidad de medida.
2. Encuentra la función que relaciona el tamaño de la entrada con el tiempo de ejecución del algoritmo, para esto utiliza la función *polyfit* de *Numpy*.
3. Calcula el tiempo de ejecución para una entrada de $1.500$ elementos.
4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño de entrada que haría que el tiempo de ejecución sea de $50$ milisegundos.


1.- La variable dependiente es la del tiempo de ejecución (ms) y la independiente es la cantidad de elementos (un)

2. Encuentra la función que relaciona el tamaño de la entrada con el tiempo de ejecución del algoritmo, para esto utiliza la función *polyfit* de *Numpy*.


```python
import numpy as np
elementos= np.array([100, 200,500,1000,2000])
tiempo=np.array([2,4,10,20,40])
m,n = np.polyfit(elementos, tiempo, 1)
print(f"la pendiente es {round(m,2)} y el coeficiente de posicion es {round(n)}")
```

La función que modela el timepo de ejecución es T(c)=0.02c

3. Calcula el tiempo de ejecución para una entrada de $1.500$ elementos.


```python
def T(c):
    return 0.02*c
```

```python
a=1500
print(f"El tiempo de ejecucion para 1500 elementos es de {T(a)} milisegundos")
```

4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño de entrada que haría que el tiempo de ejecución sea de $50$ milisegundos.


```python
from scipy.optimize import fsolve
def g(x):
    return T(x)-50
xo = np.linspace(2000,3000,1)
solucion = fsolve(g, xo)
print(solucion)
```

2. Utilizando la función *polyfit* de *Numpy* encuentra la función que modela el tiempo de transferencia $T(x)$ en función del volumen de los datos transferidos.


```python
import numpy as np
Datost= np.array([5, 10,25,50,100])
TiempoT=np.array([10,20,50,100,200])
d,t = np.polyfit(Datost, TiempoT, 1)
print(f"La pendiente es {round(d)} y el coef de posición es {round(t,)}")
```

Funcion es de T(x)=2x

3. Interprete la pendiente en el contexto del problema.


La pendiente es m= 2 lo que significa que el tiempo de transferencia aumenta 2 min por cada gb que aumentan los datos transferidos


4. ¿Cuánto tiempo toma una transferencia de $73,2$ GB?

```python

def Tr(x):
    return 2*x
```

```python
g=73.2
Tr(g)
print(f"El tiempo que demora una transferencia de 73.2GB es de {Tr(g)}")
```

5. Utiliza la función *fsolve* de *Scipy* para determinar la cantidad de datos que se puedan transferir en $123,5$ minutos.

```python
from scipy.optimize import fsolve
def A(x):
    return Tr(x)-123.5
xo = np.linspace(0,500,1)
solucion = fsolve(A, xo)
print(f"La cantidad de datos que se puedan transferir en 123.5 min es de {solucion}")
```

1. Encuentra las funciones que modelan el tiempo de carga $f(x)$ y el tiempo de descarga $g(x)$ en función del tamaño del archivo $x$. Utiliza la función *polyfit* de *Numpy* (debes encontrar dos funciones).

```python
import numpy as np
tamaño= np.array([50, 100, 250 ,500, 1000])
tiempo_carga=np.array([10, 20, 50,100,200])
tiempo_descarga=np.array([8, 16, 40,80, 160])
#para la f (tiempo de carga)
mc, nc = np.polyfit(tamaño, tiempo_carga, 1)
print(f"la pendiente es de {round(mc,2)} y el coeficiente de posicion e de {round(nc,2)}")
mc, nd = np.polyfit(tamaño, tiempo_descarga, 1)
print(f"la pendiente es de {round(mc,2)} y el coeficiente de posicion e de {round(nd,2)}")
```

Las funciones son:

CARGA: 0.2X

DESCARGA: 0.16X

```python
def f(x):
    return 0.2*x

def g(x):
    return 0.16*x
```

6. Grafica ambas funciones en el mismo gráfico, indicando nombre de los ejes y unidades de medida. Utiliza la librería *Matplotlib* para graficar.

```python
import matplotlib.pyplot as plt
x = np.arange(0,1000,1)
plt.plot(x, f(x), label = 'f(x)=tiempo de carga')
plt.plot(x, g(x), label = 'g(x)=tiempo de descarga')
plt.axis([-10,1000,-5,210])
plt.title('Relacion entre tamaño archivo y tiempo de carga y descarga')
plt.xlabel('x : Tamaño de archivo')
plt.ylabel('f(x): tiempo de carga(s), tiempo de descarga(s)')
plt.grid(True)
plt.legend()
plt.show()
```

2. Define variable dependiente e independiente, para ambas funciones, indicando su unidad de medida.

independiente x: Tamaño del archivo (mb)

Tiepo de descarga $g(x)$:

dependiente g(x): tiempo de descarga(s)

independiente x: tamaño del archivo mb

3. Calcula el tiempo de carga y descarga para un archivo de $750$ MB.

```python
a=750
print(f"Para un archivo de {a} MB el tiempo de carga es {f(a)} segundos y el de descarga {g(a)} segundos")
```

4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño del archivo si el tiempo de carga es de $163$ segundos.

```python
def L(x):
    return f(x)-163

#Valor/es inicial/es de la aproximación
xo = np.linspace(600,800,1)
solucion = fsolve(L, xo)
print(solucion)
```
